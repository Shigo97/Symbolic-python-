```python
from sympy import *
init_printing()
t = symbols('t')
```

1. A switching function $f(t)$ is defined by $f(t) = \Big\{ \begin{matrix} -3 & -2 < t < 0 \\ 3 & 0 < t < 2 \end{matrix}$ and is periodic with period 4. Sketch a graph of the function over the interval $− 4 ≤ t ≤ 4$ and find the Fourier series representing $f(t)$.

```python
Eq1 = 6*Heaviside(t)-3
```

```python
plot(Eq1, (t, -2, 2))
```

```python
s1 = fourier_series(Eq1, (t, -2, 2))
s1
```

```python
s1.truncate(5)
```

```python
plot(s1.truncate(100), (t, -4, 4))
```

3. A ramp function, $f ( t )$ , is defined as $f ( t ) = t − \pi < 0 ≤ \pi $. $f ( t )$ is periodic with period $2 \pi$.
Sketch a graph of the function over the interval $− 2 \pi ≤ t ≤ 2 \pi$ and find the Fourier series for $f ( t )$ .

```python
Eq2 = t
plot(Eq2, (t, -pi, pi))
```

```python
s2 = fourier_series(Eq2, (t, -pi, pi))
s2
```

```python
s2.truncate(5)
```

```python
plot(s2.truncate(100), (t, -2*pi, 2*pi))
```

4. A periodic function $f ( t )$ with period $2 \pi$ is defined as $f ( t ) = t$ for $− \pi < t < \pi$.
Sketch the graph of $f ( t )$ and obtain its Fourier expansion.  
You are given that $$\int t^2 \cos(nt) \, dt = \frac{2t \cos(nt)}{n^2} + \left( \frac{t^2}{n} - \frac{2}{n^3}\sin(nt) \right)$$
where $n$ is a positive integer.

```python
Eq3 = t**2
plot(Eq3, (t, -pi, pi))
```

```python
s3 = fourier_series(Eq3, (t, -pi, pi))
s3
```

```python
s3.truncate(5)
```

```python
plot(s3.truncate(100), (t, -2*pi, 2*pi))
```

5. A sawtooth function $f ( t )$ is defined by $f(t) = \Big\{ \begin{matrix} -t & -1 < t < 0 \\ t & 0 < t < 1 \end{matrix}$ and is repeated periodically.  
Sketch a graph of the function over the interval $− 2 ≤ t ≤ 2$ and find the Fourier series representing $f ( t )$ .

```python
Eq4 = abs(t) 
plot(Eq4, (t, -1, 1))
```

```python
s4 = fourier_series(Eq4, (t, -1, 1))
s4
```

```python
s4.truncate(5)
```

```python
plot(s4.truncate(15), (t, -2, 2))
```
