Find grad $\phi$ for:  
  $$\phi = x^2 - 3y$$
  
Suppose $f(x,y) = x^2 - 3y$, then:  
*  $x$ is $R[0]$ and $y$ would be $R[1]$
*  the unit vectors $\hat{i}$ and $\hat{j}$ of the 2 axes, would be respectively $R$.$x$ and $R$.$y$.

```python
from sympy import *
x, y, z = symbols('x y z')
```

```python
from sympy.physics.vector import *
R = ReferenceFrame('R')

f1 = R[0]**2 - 3*R[1]

gradient(f1, R)
```

Hence, grad$\phi = 2x\hat{i} - 3\hat{j}$  
  
To plot it you need to convert the above result into 2 separate functions; u and v.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(15,10))

x, y = np.meshgrid(np.arange(-10, 10, 1),
                      np.arange(-10, 10, 1))

u = 2*x
v = -3

plt.quiver(x, y, u, v)

plt.show()
```

Find the divergence of the following vector field:
$$ \hat{F} = x^2 \hat{i} + y^2 \hat{j} + z^2 \hat{k}$$
Suppose $F(x,y,z) = x^2 \hat{i} + y^2 \hat{j} + z^2 \hat{k}$, then:  
*  $x$ is $R[0]$, $y$ would be $R[1]$ and $z$ is $R[2]$
*  the unit vectors $\hat{i}$, $\hat{j}$, $\hat{k}$ of the 3 axes, would be respectively $R$.$x$, $R$.$y$ and $R$.$z$.

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(15,10))
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.arange(-10, 10, 2),
                      np.arange(-10, 10, 2),
                      np.arange(-10, 10, 2))

u = x**2
v = y**2
w = z**2

ax.quiver(x, y, z, u, v, w, length=0.01)

plt.show()
```

```python
F2 = R[0]**2 * R.x + R[1]**2 * R.y + R[2]**2 * R.z
F2
```

```python
divergence(F2, R)
```

Hence, div$\hat{F}$ = 2x + 2y + 2z$

Find curl$\hat{v}$ for the following two-dimensional vector fields:  
a)  $ \hat{v} = x\hat{i} + 2\hat{j} $

```python
fig = plt.figure(figsize=(15,10))

x, y = np.meshgrid(np.arange(-10, 10, 1),
                      np.arange(-10, 10, 1))

u = x
v = 2

plt.quiver(x, y, u, v)

plt.show()
```

```python
F3 = R[0]*R.x + 2*R.y
F3
```

```python
curl(F3, R)
```

b) $ \hat{v} = -y\hat{i} + x\hat{j} $

```python
fig = plt.figure(figsize=(15,10))

x, y = np.meshgrid(np.arange(-10, 10, 1),
                      np.arange(-10, 10, 1))

u = -y
v = x

plt.quiver(x, y, u, v)

plt.show()
```

```python
F4 = -R[1]*R.x + R[0]*R.y
F4
```

```python
curl(F4, R)
```

Find the rate of change of the function $4x^3 y^3 z^3$ at the point $(2.1, 1.1, 1.8)$ in the direction of the vector $(-3, 1, 2)$.

```python
f5 = 4*R[0]**3*R[1]**3*R[2]**3
f5
```

```python
F5 = gradient(f5,R)
F5
```

```python
numF5 = (F5.subs({R[0]:2, R[1]:1.1, R[2]:1.8}))
numF5
```

```python
vec2 = -3*R.x + 1*R.y + 2*R.z
vec2
```

```python
vec2Norm = vec2.normalize()
vec2Norm
```

```python
numF5.dot(vec2Norm).evalf()
```
